```python
covid_data=pd.read_csv('C:\\Users\\nazeer\\Downloads\\StatewiseTestingDetails.csv')
```

```python
covid_data_by_State = covid_data.groupby('State')[['TotalSamples','Positive' ,'Negative']].sum().reset_index()
covid_data_by_State
```

```python

plt.figure(figsize=(10,6))

plt.title("TotalSamples")
sns.barplot(x=covid_data_by_State.index, y=covid_data_by_State['TotalSamples'])
plt.ylabel("TotalSamples")
```

```python
# visualisation with positive sample
plt.figure(figsize=(10,6))
plt.title("Positive Samples")
sns.barplot(x=covid_data_by_State.index, y=covid_data_by_State['Positive'])
plt.ylabel("Positive Samples")
```
